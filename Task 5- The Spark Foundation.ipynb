### Top 10 batsman 

```python
top_10_batsman=score_data.groupby('batsman')['batsman_runs'].sum().nlargest(10).reset_index()
top_10_batsman_names=top_10_batsman['batsman'].tolist()
top_10_batsman_score=top_10_batsman['batsman_runs'].tolist()
top_10_batsman_and_score=pd.DataFrame(list(zip(top_10_batsman_names, top_10_batsman_score)),columns=['batsman_name','scores'])
top_10_batsman_and_score
```

```python
plt.subplots(figsize=(20,10))
sns.barplot(x=top_10_batsman_names, y=top_10_batsman_score, palette='flare')
plt.title('top 10 batsman', fontsize=25)
plt.xlabel('name of batsman', fontsize=20)
plt.ylabel('runs', fontsize=20)
plt.xticks(rotation=50)
plt.show()
```

### Batsman with most number of sixes

```python
df=score_data.groupby('batsman')['batsman_runs'].agg(lambda x: (x==6).sum()).reset_index().sort_values(by='batsman_runs', ascending=False)
df=df.iloc[:10,:]
labels=np.array(df['batsman'])
ind=np.arange(len(labels))
width=0.9
fig,ax=plt.subplots(figsize=(15,8))
rects=ax.bar(ind,np.array(df['batsman_runs']),width=width, color='green')
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels,rotation=90)
ax.set_ylabel('count')
ax.set_title('batsman with most number of sixes')
ax.set_xlabel('batsman name')
plt.show()
```

### Batsman with number of dot balls

```python
df=score_data.groupby('batsman')['batsman_runs'].agg(lambda x: (x==0).sum()).reset_index().sort_values(by='batsman_runs', ascending=False)
df=df.iloc[:10,:]
labels=np.array(df['batsman'])
ind=np.arange(len(labels))
width=0.9
fig,ax=plt.subplots(figsize=(15,8))
rects=ax.bar(ind,np.array(df['batsman_runs']),width=width, color='orange')
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels,rotation=90)
ax.set_ylabel('count')
ax.set_title('batsman with most number of sixes')
ax.set_xlabel('batsman name')
plt.show()
```

###  batsman with highest % of dot balls

```python
def balls_faced(x):
    return len(x)

def dot_balls(x):
    return(x==0).sum()

df=score_data.groupby('batsman')['batsman_runs'].agg([balls_faced, dot_balls]).reset_index()
df=df.loc[df.balls_faced>200,:]
df['percentage_of_dot_balls']=(df['dot_balls']/df['balls_faced'])*100.
df=df.sort_values(by='percentage_of_dot_balls', ascending=False).reset_index(drop=True)
df=df.iloc[:10,:]

fig, ax1= plt.subplots(figsize=(15,8))
ax2=ax1.twinx()
labels=np.array(df['batsman'])
ind=np.arange(len(labels))
width=0.9
rects=ax1.bar(ind, np.array(df['dot_balls']),width=width, color='cyan')
ax1.set_xticks(ind+((width)/2.))
ax1.set_xticklabels(labels, rotation='vertical')
ax1.set_ylabel('count of dot balls', color='black')
ax1.set_title('batsman with highest percentage of dot balls(balls faced>200)')
ax2.plot(ind+0.45, np.array(df['percentage_of_dot_balls']), color='b', marker='o')
ax2.set_ylabel('percentage of dot balls', color='b')
ax2.set_ylim([0,100])
ax2.grid(visible=False)
plt.show()
```

### Bowler Analysis

### Bowler who has balled most number of balls in IPL

```python
df=score_data.groupby('bowler')['ball'].agg('count').reset_index().sort_values(by='ball', ascending=False).reset_index()
df=df.iloc[:10,:]

labels=np.array(df['bowler'])
ind=np.arange(len(labels))
width=0.9
fig,ax=plt.subplots(figsize=(15,8))
rects=ax.bar(ind, np.array(df['ball']), width=width, color='magenta')
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels, rotation='vertical')
ax.set_ylabel('count')
ax.set_title('top bowlers- number of balls bowled in IPL')
ax.set_xlabel('bowler_names')
plt.show()
```

### top bowlers- number of dot balls bowled in IPL

```python
df=score_data.groupby('bowler')['total_runs'].agg(lambda x: (x==0).sum()).reset_index().sort_values(by='total_runs', ascending=False)
df=df.iloc[:10,:]
labels=np.array(df['bowler'])
ind=np.arange(len(labels))
width=0.9
fig,ax=plt.subplots(figsize=(15,8))
rects=ax.bar(ind,np.array(df['total_runs']),width=width, color='lightskyblue')
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels,rotation='vertical')
ax.set_ylabel('count')
ax.set_title('top bowlers- number of dot balls bowled in IPL')
ax.set_xlabel('bowler name')
plt.show()
```

### bowlers who bowled more number of extras in IPL

```python
df=score_data.groupby('bowler')['extra_runs'].agg(lambda x: (x>0).sum()).reset_index().sort_values(by='extra_runs', ascending=False)
df=df.iloc[:10,:]

labels=np.array(df['bowler'])
ind=np.arange(len(labels))
width=0.9
fig,ax=plt.subplots(figsize=(15,8))
rects=ax.bar(ind,np.array(df['extra_runs']),width=width, color='yellow')
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels,rotation='vertical')
ax.set_ylabel('count')
ax.set_title('bowlers with more extras in IPL')
ax.set_xlabel('bowler name')
plt.show()
```
